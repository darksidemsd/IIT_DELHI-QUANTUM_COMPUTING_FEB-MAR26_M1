```python
from scipy.stats import beta
import numpy as np

def clopper_pearson(k, n, alpha=0.05):
    lower = beta.ppf(alpha/2, k, n-k+1)
    upper = beta.ppf(1-alpha/2, k+1, n-k)
    return lower, upper

true_p = 0.3
shots_list = [200,1000,5000]

for N in shots_list:
    k = np.random.binomial(N, true_p)
    p_hat = k/N
    low, high = clopper_pearson(k,N)
    print("\nShots:",N)
    print("Estimate:",p_hat)
    print("95% CI:", (low,high))
    print("Width:", high-low)
```
