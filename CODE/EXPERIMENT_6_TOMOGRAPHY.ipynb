```python
import numpy as np
```

```python
from qiskit.quantum_info import Statevector

theta = np.pi/4
phi = np.pi/6

alpha = np.cos(theta/2)
beta = np.exp(1j*phi)*np.sin(theta/2)

true_state = Statevector([alpha, beta])

# Pauli matrices
sigma_x = np.array([[0,1],[1,0]])
sigma_y = np.array([[0,-1j],[1j,0]])
sigma_z = np.array([[1,0],[0,-1]])

x = np.real(true_state.expectation_value(sigma_x))
y = np.real(true_state.expectation_value(sigma_y))
z = np.real(true_state.expectation_value(sigma_z))

print("Bloch Vector:", (x,y,z))
```
