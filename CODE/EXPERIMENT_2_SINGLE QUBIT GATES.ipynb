```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import numpy as np

def analyze_circuit(qc):
    sv = Statevector.from_instruction(qc)
    probs = sv.probabilities()
    x = np.real(sv.expectation_value([[0,1],[1,0]]))
    y = np.real(sv.expectation_value([[0,-1j],[1j,0]]))
    z = np.real(sv.expectation_value([[1,0],[0,-1]]))
    return sv, probs, (x,y,z)

# Example Gates
gates = ["x","h","z"]

for g in gates:
    qc = QuantumCircuit(1)
    getattr(qc, g)(0)
    sv, probs, bloch = analyze_circuit(qc)
    print("\nGate:", g.upper())
    print("State:", sv)
    print("Probabilities:", probs)
    print("Bloch:", bloch)
```

```python
qc = QuantumCircuit(1)
qc.rx(np.pi/2, 0)
sv, probs, bloch = analyze_circuit(qc)
print("\nRX(pi/2)")
print("Probabilities:", probs)
print("Bloch:", bloch)
```
