```python
import numpy as np

# Define states
states = {
    "|0>": np.array([1, 0]),
    "|1>": np.array([0, 1]),
    "|+>": np.array([1/np.sqrt(2), 1/np.sqrt(2)]),
    "|->": np.array([1/np.sqrt(2), -1/np.sqrt(2)]),
    "|+i>": np.array([1/np.sqrt(2), 1j/np.sqrt(2)]),
    "|-i>": np.array([1/np.sqrt(2), -1j/np.sqrt(2)])
}

def bloch_vector(state):
    alpha, beta = state
    theta = 2*np.arccos(abs(alpha))
    phi = np.angle(beta) - np.angle(alpha)

    x = np.sin(theta)*np.cos(phi)
    y = np.sin(theta)*np.sin(phi)
    z = np.cos(theta)

    return np.real(x), np.real(y), np.real(z), theta, phi

for name, state in states.items():
    alpha, beta = state
    P0 = abs(alpha)**2
    P1 = abs(beta)**2
    x,y,z,theta,phi = bloch_vector(state)

    print("\nState:", name)
    print("α =", alpha, " β =", beta)
    print("P(0) =", P0, " P(1) =", P1)
    print("θ =", theta, " φ =", phi)
    print("Bloch vector =", (x,y,z))
```
